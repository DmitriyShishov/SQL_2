## Выполнение задания

Будем работать с копией датасета df_chat

```python
df = df_chat.copy()
```

```python
df[(df['created_by'] == 0) & (df['type'] == 'outgoing_chat_message')]
```

У нас выявлено 173 сообщения, которые помечены, как отправленным менеджером, но с идентификатором 0. Это явный сбой в системе. 
Так как в условии ясно сказано, что идентификатор 0 всегда у клиента, я решил в этих 173 строках изменить тип на входящий, отправленный клиентом менеджеру

```python
df.type.value_counts()
```

```python
df['type'] = df.apply(lambda x: 'incoming_chat_message' if (x.created_by == 0) & (x.type == 'outgoing_chat_message') else x.type, axis=1)
df.type.value_counts()
```

Далее переведем время сообщения из формата UnixTimestamp в тип Даты и времени с учетом часового пояса

```python
df.loc[:, ['created_at']] = df['created_at'].apply(lambda x: pd.Timestamp(x, unit='s', tz='Europe/Moscow'))
df.head()
```

Затем отсортируемся по номеру сделки и времени для лучшего понимания

```python
df = df.sort_values(by=['entity_id', 'created_at'])
df.head(20)
```

Для следующих действий необходимо получить дополнительные столбцы: Время и тип сообщения на предыдущей строке.

```python
df['previous_type'] = df.groupby('entity_id')['type'].shift()
df['previous_time'] = df.groupby('entity_id')['created_at'].shift()
df.head(20)
```

Теперь на основе созданных столбцов мы можем получить новый признак с времени исходящего сообщения для каждого ответа. При этом мы учитываем, чтобы значения создавались только для исходящий сообщений и тип сообщения не был одинаков на двух и более строках подряд. Это поможет нам учесть случаи, когда несколько подряд сообщений могут идти от одного адресата.
В конце удалим вспомогательные столбцы

```python
df['question_time'] = df.apply(lambda x: x['previous_time'] if (x['type'] != x['previous_type']) & (x['type'] == 'outgoing_chat_message') else None, axis=1)
df = df.drop(['previous_type', 'previous_time'], axis=1)
df.head(20)
```

Далее я хотел продемонстрировать исходящие ответы менеджеров, которые были отправлены с полуночи до начала рабочего времени.

```python
df[(df['created_at'].dt.time.between(pd.Timestamp('00:00:00').time(), pd.Timestamp('09:30:00').time())) & (df['type'] == 'outgoing_chat_message') & (df['question_time'].notna())]
```

Тут встречаются и случаи, когда менеджер пришел на работу раньше и ответил на сообщение, когда менеджер задержался на работе и ответил после полуночи, а есть случаи, когда менеджер отвечал ночью. Разницы во времени ответа во всех этих случаях сильно разнятся, поэтому, так как в целом расчет среднего всегда чувствителен к аномальным значениям, я решил все эти случаи убрать из выборки.

Также мы убираем все входящие сообщения и исходящие, для которых не было вопроса от клиента(сделка состоит из одного исходящего сообщения)

```python
df = df[(df['question_time'].notna()) & (df['type'] == 'outgoing_chat_message') & (~df['created_at'].dt.time.between(pd.Timestamp('00:00:00').time(), pd.Timestamp('09:30:00').time()))]
```

Далее определим функцию расчета разницы между временем вопроса и временем ответа.

1) Определем дату и время вопроса(q_time) и ответа(a_time);
2) Определяем переменную work_time_start, которяа будет определять начало рабочего дня и будем изменять дату в соответствии с датой ответа;
3) Для случая, когда вопрос был получен с полуночи до начала рабочего дня, разница будет определяться между временем ответа и временем начала рабочего дня;
4) Для случаев, когда вопрос был получен до полуночи, а ответ был отправлен утром после начала рабочего дня, разница будет считаться между временем ответа и вопроса за исключением 9 часов 30 минут, когда менеджер не находился на рабочем месте.
5) Во всех остальных случаях мы просто получаем разницу между временем ответа и временем вопроса

```python
def calculate_time_diff(row):
    q_time = row['question_time']
    a_time = row['created_at']

    work_time_start = a_time.normalize() + pd.Timedelta(hours=9, minutes=30)

    if (q_time >= a_time.normalize()) and (q_time < work_time_start) and (a_time >= work_time_start):
        return a_time - work_time_start
    elif q_time < a_time.normalize() and a_time >= work_time_start:
        return a_time - q_time - pd.Timedelta(hours=9, minutes=30)
    else:
        return a_time - q_time
```

```python
df['time_diff'] = df.apply(calculate_time_diff, axis=1)
df.head(20)
```

В качестве примеров корректной работы функции посмотрим несколько случаев

```python
df[df['entity_id'] == 36698385].iloc[[0], :]
```

Здесь мы видим, что вопрос был получен в 08:19 утра, а ответ отправлен в 10:58. В таком случае разница должна считается от момента начала рабочего дня, и равняться 01:28:00. Так и есть.

Другой пример:

```python
df[df['entity_id'] == 36827109].iloc[[0], :]
```

Здесь вопрос был получен до завершения рабочего дня(в 23:22:32), а ответ отправлен в 09:52:22 утром следующего дня. Как видим, функция здесь также отрабатывает корректно и вместо 10 с половиной часов разницы мы получаем разницу около часа.

Теперь переведем полученное значения разницы в минуты для лучшего восприятия. Для этого сначала переведем тип в формат секунд, возьмем их целое количесвто и разделим на 60

```python
df['time_diff_min'] = round(df['time_diff'].astype('timedelta64[s]').astype('int64') / 60, 2)

df.head(20)
```

Теперь мы можем рассчитать среднее количество минут на ответ каждого менеджера. Группировку производим по столбцу created_by.

```python
df['avg_time_diff_min'] = round(df.groupby('created_by')['time_diff_min'].transform('mean'), 1)
df.head(20)
```

Теперь скомпануем итоговую сводную таблицу. Сначала соединим текущую таблицу с датасетом, содержащим информацию о менеджерах. Выберем только нужные столбцы и уберем дубликаты.

```python
df_final = pd.merge(df, df_managers, left_on='created_by', right_on='mop_id', how='right')[['mop_id', 'name_mop', 'rop_id', 'avg_time_diff_min']].drop_duplicates()
df_final
```

Теперь добавим в финальную сводную таблицу имя руководителя каждого менеджера. Снова используем метод merge. В конце отсортируемся по среднему значению ответа

```python
df_final = pd.merge(df_final, df_rops, on='rop_id', how='inner')[['mop_id', 'name_mop', 'rop_name', 'avg_time_diff_min']].sort_values('avg_time_diff_min')
df_final
```
